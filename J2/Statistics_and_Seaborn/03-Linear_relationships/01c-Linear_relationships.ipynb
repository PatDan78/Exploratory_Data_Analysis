1. Import usual librairies

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("IBM_HR_ATTRITION.csv")
```

2. We would like to know the probability of leaving the business in relation to the distance from home. Use _lmplot()_ to view this. What can you conclude?

Nous aimerions connaître la probabilité de quitter l'entreprise par rapport à l'éloignement du domicile. Utilisez lmplot() pour voir ceci. Que pouvez-vous conclure ?


```python
# Compter les valeurs unique dans la colonne "Attrition"

["Attrition"].unique()
```

```python
# Renvoie une série contenant la fréquence de chaque ligne distincte du Dataframe.

dataset['Attrition'].value_counts()
```

```python
# Transforme les éléments str en int64

dataset['Attrition'] = dataset['Attrition'].apply(lambda x: 1 if x.lower()=="yes" else 0)
```

```python
dataset['Attrition']
```

```python
# rapport entre "Attrition" et "DistanceFromHome"

sns.lmplot(x="DistanceFromHome", y="Attrition", data= dataset, logistic = True, scatter_kws={"alpha":.05})
```

```python
sns.lmplot(x="DistanceFromHome", y="Attrition", data= dataset, logistic = True, scatter_kws={"alpha":.05})
```
